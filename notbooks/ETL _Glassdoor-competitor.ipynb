## Scraping des données

```python
ratings =[]
working_dates =[]
review_headlines =[]
review_links =[]
author_job_titles =[]
review_pros =[]
review_cons =[]
def get_glassdoor_posts (soup):
    containers = soup.find_all(class_='gdReview')
    for container in containers :
        try :
            rating = container.find('span')
            rating = rating.get_text()
            ratings.append(rating)

        except :
            ratings.append('none')

        try :
            working_date = container.find(class_='pt-xsm pt-md-0 css-1qxtz39 eg4psks0')
            working_dates.append(working_date.get_text())

        except :
            working_dates.append('none')
        try:
            review_headline = container.find(class_='reviewLink')
            review_headlines.append(review_headline.get_text())
        except :
            review_headlines.append('none')

        try :
            review_link = container.find(class_='mb-xxsm mt-0 css-93svrw el6ke055')
            review_links.append(review_link.a['href'])
        except:
            review_links.append('none')
        try :
            author_job_title = container.find(class_='authorJobTitle middle common__EiReviewDetailsStyle__newGrey')
            author_job_titles.append(author_job_title.get_text())
        except :
            author_job_titles.append('none')
        try :
            review_pro = container.find(class_='mt-0 mb-0 pb v2__EIReviewDetailsV2__bodyColor v2__EIReviewDetailsV2__lineHeightLarge v2__EIReviewDetailsV2__isCollapsed')
            review_pros.append(review_pro.get_text())
        except :
            review_pros.append('none')
        try :
            review_con = container.find_all(class_='mt-0 mb-0 pb v2__EIReviewDetailsV2__bodyColor v2__EIReviewDetailsV2__lineHeightLarge v2__EIReviewDetailsV2__isCollapsed')
            review_cons.append(review_con[1].get_text())
        except :
            review_cons.append('none')
```

## Scrolling à la fin de page

```python
def scrolling_function():
    start = time.time()
    initialScroll = 0
    finalScroll = 10
  
    while True:
        driver.execute_script(f"window.scrollTo({initialScroll},{finalScroll})")
        initialScroll = finalScroll
        finalScroll += 10
        end = time.time()
        if round(end - start) > 3:
            break 
```

## traitement de donnée 

```python
def cleantext2(text):
    text = re.sub(r'@[A-Za-z0-9]+|{|}|(|)|-|!', '', str(text))
    return text
Reviews_Data['review_headlines']= Reviews_Data['review_headlines'].apply(cleantext2)
```

```python
review_list=[]
review_list =list(Reviews_Data['review_headlines'])
```

 ## Appliquer le modèle distilbert 

```python
from transformers import pipeline 
from transformers import AutoTokenizer, AutoModelForSequenceClassification
Model_Name = ('distilbert-base-uncased-finetuned-sst-2-english')#exporta baad 7adher modele mtaaou exporta
model =  AutoModelForSequenceClassification.from_pretrained(Model_Name)
tokenizer=AutoTokenizer.from_pretrained(Model_Name)
classifier = pipeline('sentiment-analysis',model =model, tokenizer=tokenizer)
results=classifier(review_list)
```

```python
results
```

```python
import pandas as pd
p =0
n=0
for i in range (0,len(results)):
    if results[i]['label']=='POSITIVE':
        p+=1 
    else:
        n+=1

data1 = [['inetum',round((p/128)*100,2),round((n/128)*100,2)]]
df1 = pd.DataFrame(data1, columns = ['competitor', 'postive','negative'])
```
