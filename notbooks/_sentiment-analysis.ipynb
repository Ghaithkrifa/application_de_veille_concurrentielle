```python
from hdfs import InsecureClient
import pandas as pd
import re
from transformers import pipeline 
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
file_name= "accenture/ACCENTURE_Glassdoor_Reviews_df"
hdfs_url = "http://localhost:9870/"
hdfs_user = "krifa"
client = InsecureClient(hdfs_url, user=hdfs_user)
with client.read(file_name, encoding = 'utf-8') as reader:
  df =  pd.read_csv(reader)
```

```python
from textblob import TextBlob
from wordcloud import WordCloud, STOPWORDS
```

```python
def cleantext2(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', str(text))
    

    return text
df['review_headlines']= df['review_headlines'].apply(cleantext2)
```

```python
df['review_headlines']
```

```python
review_list=[]
review_list =list(df['review_headlines'])
```

```python
Model_Name = ('distilbert-base-uncased-finetuned-sst-2-english')
model =  AutoModelForSequenceClassification.from_pretrained(Model_Name)
tokenizer=AutoTokenizer.from_pretrained(Model_Name)
```

```python
classifier = pipeline('sentiment-analysis',model =model, tokenizer=tokenizer)
```

```python
results=classifier(review_list)
```

```python
results
```

```python
p =0
n=0
for i in range (0,len(results)):
    if results[i]['label']=='POSITIVE':
        p+=1 
    else:
        n+=1
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

#create two new clumns
df['Subjectivity'] = df ['review_headlines'].apply(getSubjectivity)
df['Polarity'] = df ['review_headlines'].apply(getPolarity)
```

```python
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
df['Analysis'] = df['Polarity'].apply(getAnalysis)
```
